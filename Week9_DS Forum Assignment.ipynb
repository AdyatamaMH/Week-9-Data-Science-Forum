```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
iris_data = pd.read_csv('iris.csv')
iris_data.head()
```

```python
# Load Iris dataset
iris_data = pd.read_csv('iris.csv')

# Separate features from the dataset
X = iris_data.drop('species', axis=1)
```

```python
# KMeans clustering
kmeans = KMeans(n_clusters=3)
kmeans_labels = kmeans.fit_predict(X)

plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
sns.scatterplot(x='petal_length', y='petal_width', data=iris_data, hue=kmeans_labels, palette='viridis')
plt.title('KMeans Clustering')

plt.tight_layout()
plt.show()
```

```python
# DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Visualization of results
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 2)
sns.scatterplot(x='petal_length', y='petal_width', data=iris_data, hue=dbscan_labels, palette='viridis')
plt.title('DBSCAN Clustering')

plt.tight_layout()
plt.show()
```

```python
# Agglomerative Clustering
agg_cluster = AgglomerativeClustering(n_clusters=3)
agg_labels = agg_cluster.fit_predict(X)

plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 3)
sns.scatterplot(x='petal_length', y='petal_width', data=iris_data, hue=agg_labels, palette='viridis')
plt.title('Agglomerative Clustering')

plt.tight_layout()
plt.show()
```

```python
# Gaussian Mixture Models (GMM)
gmm = GaussianMixture(n_components=3)
gmm_labels = gmm.fit_predict(X)

plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 4)
sns.scatterplot(x='petal_length', y='petal_width', data=iris_data, hue=gmm_labels, palette='viridis')
plt.title('Gaussian Mixture Model')

plt.tight_layout()
plt.show()
```

```python
# KMeans clustering
kmeans = KMeans(n_clusters=3)
kmeans_labels = kmeans.fit_predict(X)
kmeans_silhouette = silhouette_score(X, kmeans_labels)
kmeans_davies_bouldin = davies_bouldin_score(X, kmeans_labels)

# DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
dbscan_silhouette = silhouette_score(X, dbscan_labels)
dbscan_davies_bouldin = davies_bouldin_score(X, dbscan_labels)

# Agglomerative Clustering
agg_cluster = AgglomerativeClustering(n_clusters=3)
agg_labels = agg_cluster.fit_predict(X)
agg_silhouette = silhouette_score(X, agg_labels)
agg_davies_bouldin = davies_bouldin_score(X, agg_labels)

# Gaussian Mixture Models (GMM)
gmm = GaussianMixture(n_components=3)
gmm_labels = gmm.fit_predict(X)
gmm_silhouette = silhouette_score(X, gmm_labels)
gmm_davies_bouldin = davies_bouldin_score(X, gmm_labels)

# Displaying scores
print("Clustering Methods    :", "kmeans_labels,", "agglomerative_labels,", "dbscan_labels,", "gmm_labels")
print("Silhouette Scores     :", [kmeans_silhouette, agg_silhouette, dbscan_silhouette, gmm_silhouette])
print("Davies Bouldin Indices:", [kmeans_davies_bouldin, agg_davies_bouldin, dbscan_davies_bouldin, gmm_davies_bouldin])
```
